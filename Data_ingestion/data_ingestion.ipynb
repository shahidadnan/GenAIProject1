*Web Based Loader*

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader=WebBaseLoader(web_paths=("https://kinginstitute.stanford.edu/malcolm-x",),
                     bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_=("jumpstart-ui--two-column flex-container centered-container"))))
```

```python
webdata=loader.load()
```

```python
from langchain_community.document_loaders import WikipediaLoader
docs = WikipediaLoader(query="Malcolm X", load_max_docs=2).load()
docs
```
